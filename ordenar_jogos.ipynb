```python
import pandas as pd
import locale

```

**Selecionar a tabela e corrigir os dados numéricos**

```python
url = fn
tables = pd.read_html(url) 

sorteios = tables[0]

sorteios['Rateio_Sena'] = sorteios['Rateio_Sena'].apply(lambda x: x.replace('.',''))
sorteios['Rateio_Sena'] = sorteios['Rateio_Sena'].apply(lambda x: x.replace(',','.'))
sorteios['Rateio_Sena'] = sorteios['Rateio_Sena'].apply(lambda x: float(x))

sorteios['Rateio_Quina'] = sorteios['Rateio_Quina'].apply(lambda x: x.replace('.',''))
sorteios['Rateio_Quina'] = sorteios['Rateio_Quina'].apply(lambda x: x.replace(',','.'))
sorteios['Rateio_Quina'] = sorteios['Rateio_Quina'].apply(lambda x: float(x))

sorteios['Rateio_Quadra'] = sorteios['Rateio_Quadra'].apply(lambda x: x.replace('.',''))
sorteios['Rateio_Quadra'] = sorteios['Rateio_Quadra'].apply(lambda x: x.replace(',','.'))
sorteios['Rateio_Quadra'] = sorteios['Rateio_Quadra'].apply(lambda x: float(x))

print (sorteios.dtypes)

```

```python
sorteios.head()
```

**Ordenar pelos prêmios da sena e criar a coluna de total sena**

```python
totalizado = sorteios["Rateio_Sena"] * sorteios["Ganhadores_Sena"]
sorteios["totalizado_sena"] = totalizado
ordenado = sorteios.sort_values(by='Rateio_Sena', ascending=False)


ordenado.head()
```

**Resultados:**
1.   Mostrar o total de prêmio pago pela sena
2.   Mostrar quantas vezes o prêmio foi maior que 40 milhões
3. Quantos foram para os políticos do DF





```python
somatorio = ordenado["totalizado_sena"].sum()

print ("Total de senas pagas: R$ {:0,.2f}".format(somatorio))

maior_que_40 = ordenado.query("`totalizado_sena` >= 40000000")
premios_para_df = ordenado.query("UF == 'DF'")

quantidade_premios_altos = maior_que_40[maior_que_40.columns[0]].count()
quantidade_premios_df = premios_para_df[premios_para_df.columns[0]].count()

print("Quantidade de prêmios acima de R$40.000.000: ", quantidade_premios_altos )
print("Quantidade de prêmios acima de R$40.000.000 para o DF: ", quantidade_premios_df )
# primeira_linha = ordenado.iloc[0]

# print (totalizado)
# print (primeira_linha["Rateio_Sena"])
```
